**Task 4:**

*Suppose you are flipping a coin - each with a probability p of giving heads - plot the entropy of the total number of heads versus p.*

**Answer:**

The trick is to use numpy.random.binomial - this allows you to drawn from a binomial distribution with specified parameters, where n = trials and prob_ty = probability, p is in the interval . So in this case as a coin has only two sides our intervals are 0 or 1. 

The retun function then allows you to fetch the number of outcomes an interval occured for the number of trials.

```python
# For Fair coin flip chance of heads or tails 50/50
import numpy as np
import matplotlib.pyplot as plt 

def event(n,prob_ty,size):    
    Outcome = np.random.binomial(n,prob_ty,size)    
    return Outcome

Trials = 20
prob_ty = 0.5
Outcome = np.random.binomial(Trials,prob_ty)
print(Outcome)

Prob_expt=[np.equal(Outcome,i).mean() for i in range(Trials)]

plt.plot(Prob_expt,'r--')
plt.xlabel('Heads Returned')
plt.ylabel('Probability of Occurrence',)
plt.show()
```
